```python
#Question 1
import matplotlib.pyplot as plt
import pandas as pd
import io

data = """Year,Population
1950,2557373823
1960,3042420131
1970,3712078028
1980,4451000109
1990,5320816688
2000,6145006989
2010,6958169159
2020,7794798739
"""

df = pd.read_csv(io.StringIO(data))

plt.figure(figsize=(10, 6))

plt.plot(df['Year'], df['Population'], marker='o', linestyle='-', color='dodgerblue')

plt.xlabel('Year')
plt.ylabel('Population (in Billions)')
plt.title('World Population Growth Across Decades')

plt.grid(True)

plt.ticklabel_format(style='plain', axis='y')
plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, p: format(int(x), ','))
)

plt.show()
```

```python
#Question 2
import matplotlib.pyplot as plt
import pandas as pd
import io

data = """Date,Close
2023-01-03,125.07
2023-01-10,130.73
2023-01-17,135.94
2023-01-24,142.53
2023-01-31,144.29
2023-02-07,154.65
2023-02-14,153.20
2023-02-21,148.48
2023-02-28,147.41
2023-03-07,151.60
2023-03-14,152.59
2023-03-21,159.28
2023-03-28,160.77
"""
df = pd.read_csv(io.StringIO(data))

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12, 6))

plt.plot(df['Date'], df['Close'], color='green', linewidth=2)

plt.title('Apple (AAPL) Stock Price Trend (Q1 2023)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True, linestyle='--', alpha=0.7)

plt.gcf().autofmt_xdate()

plt.show()
```
