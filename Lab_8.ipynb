```python
#Question 1
import pandas as pd
import matplotlib.pyplot as plt
import io

data = """Subject,Average Marks
Math,75
Science,82
English,88
History,79
Art,92
"""

df = pd.read_csv(io.StringIO(data))

subjects = df['Subject']
average_marks = df['Average Marks']

plt.figure(figsize=(10, 6))

bar_colors = ['#4A90E2', '#50E3C2', '#F5A623', '#BD10E0', '#9013FE']

plt.bar(subjects, average_marks, color=bar_colors)

plt.xlabel('Subjects', fontsize=12, fontweight='bold')
plt.ylabel('Average Marks', fontsize=12, fontweight='bold')
plt.title('Comparison of Average Student Marks by Subject', fontsize=16, fontweight='bold')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Questi0on 2
import pandas as pd
import matplotlib.pyplot as plt
import io

data = """Category,Sales
Electronics,150000
Clothing,85000
Home Goods,110000
Books,45000
Groceries,130000
"""

df = pd.read_csv(io.StringIO(data))

categories = df['Category']
sales = df['Sales']

plt.figure(figsize=(10, 7))

bar_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

plt.bar(categories, sales, color=bar_colors)

plt.xlabel('Product Category', fontsize=12, fontweight='bold')
plt.ylabel('Sales (in USD)', fontsize=12, fontweight='bold')
plt.title('Total Sales by Product Category', fontsize=16, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
