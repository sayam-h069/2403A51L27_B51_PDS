```python
import pandas as pd
import matplotlib.pyplot as plt

budget_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
    'Budget': [50000, 150000, 120000, 80000, 100000]
}
budget_df = pd.DataFrame(budget_data)

expenses_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
    'Actual_Expenses': [55000, 140000, 130000, 85000, 95000]
}
expenses_df = pd.DataFrame(expenses_data)

print("\nDepartment Budget DataFrame:")
print(budget_df)
print("\nActual Expenses DataFrame:")
print(expenses_df)

merged_df = pd.merge(budget_df, expenses_df, on='Department')
print("\nMerged DataFrame:")
print(merged_df)

merged_df['Variance'] = merged_df['Budget'] - merged_df['Actual_Expenses']
print("\nDataFrame with Variance:")
print(merged_df)

exceeding_df = merged_df[merged_df['Variance'] < 0]
print("\nDepartments exceeding budget:")
print(exceeding_df)

sorted_df = merged_df.sort_values(by='Variance')
print("\nDepartments sorted by variance:")
print(sorted_df)

plt.figure(figsize=(8,5))
plt.bar(sorted_df['Department'], sorted_df['Budget'], alpha=0.6, label='Budget')
plt.bar(sorted_df['Department'], sorted_df['Actual_Expenses'], alpha=0.6, label='Actual Expenses')
plt.title('Department Budget vs Actual Expenses')
plt.xlabel('Department')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.show()
```
