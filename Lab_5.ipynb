```python
#Question-1
import pandas as pd
import numpy as np

students_data = {
    'student_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'major': ['Math', 'Science', 'Math', 'History', 'Science']
}
students_df = pd.DataFrame(students_data)

scores_data = {
    'student_id': [101, 102, 103, 106],
    'score': [85, 92, 78, 88]
}
scores_df = pd.DataFrame(scores_data)

print("--- Students Dataset ---")
print(students_df)
print("\n--- Scores Dataset ---")
print(scores_df)

inner_merged_df = pd.merge(students_df, scores_df, on='student_id', how='inner')
print("\n--- Inner Join ---")
print(inner_merged_df)

left_merged_df = pd.merge(students_df, scores_df, on='student_id', how='left')
print("\n--- Left Join ---")
print(left_merged_df)

right_merged_df = pd.merge(students_df, scores_df, on='student_id', how='right')
print("\n--- Right Join ---")
print(right_merged_df)

print("\n--- Aggregate Statistics for Inner Join ---")
mean_score = inner_merged_df['score'].mean()
max_score = inner_merged_df['score'].max()
min_score = inner_merged_df['score'].min()
score_count = inner_merged_df['score'].count()

print(f"Mean Score: {mean_score:.2f}")
print(f"Max Score: {max_score}")
print(f"Min Score: {min_score}")
print(f"Count of Scores: {score_count}")

print("\n--- Descriptive Statistics Summary ---")
print(inner_merged_df['score'].describe())

print("\n--- Grouped Statistics by Major ---")
grouped_stats = inner_merged_df.groupby('major')['score'].agg(['mean', 'max', 'min', 'count'])
print(grouped_stats)
```

```python
#question-2
import pandas as pd

customers_data = {
    'customer_id': [101, 102, 103, 104, 105, 106],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'region': ['North', 'South', 'North', 'East', 'South', 'West']
}
customers_df = pd.DataFrame(customers_data)
customers_df.to_csv('customers.csv', index=False)

sales_data = {
    'sale_id': range(1, 8),
    'customer_id': [101, 102, 103, 102, 105, 101, 105],
    'amount': [150.50, 200.00, 340.10, 55.75, 120.00, 99.99, 300.00]
}
sales_df = pd.DataFrame(sales_data)
sales_df.to_csv('sales.csv', index=False)

customers = pd.read_csv('customers.csv')
sales = pd.read_csv('sales.csv')

print("--- Customers Dataset ---")
print(customers)
print("\n--- Sales Dataset ---")
print(sales)

inner_join_df = pd.merge(customers, sales, on='customer_id', how='inner')
print("\n--- Merged Dataset (Inner Join) ---")
print(inner_join_df)

left_join_df = pd.merge(customers, sales, on='customer_id', how='left')
print("\n--- Merged Dataset (Left Join) ---")
print(left_join_df)

total_sales = inner_join_df['amount'].sum()
average_sale_value = inner_join_df['amount'].mean()

print("\n--- Overall Sales Statistics ---")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Average Sale Value: ${average_sale_value:,.2f}")

sales_per_customer = inner_join_df.groupby('name')['amount'].sum().reset_index()
print("\n--- Total Sales Per Customer ---")
print(sales_per_customer.sort_values(by='amount', ascending=False))

regional_sales = inner_join_df.groupby('region')['amount'].agg(['sum', 'mean', 'count'])
regional_sales = regional_sales.rename(columns={
    'sum': 'Total Sales',
    'mean': 'Average Sale',
    'count': 'Number of Sales'
})
print("\n--- Aggregate Sales Statistics by Region ---")
print(regional_sales.sort_values(by='Total Sales', ascending=False))
```
